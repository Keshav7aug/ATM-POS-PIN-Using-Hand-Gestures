```python
import tkinter as tk
import tkinter.font as font
import cv2
```

```python
#root.geometry("500x200")
```

```python
def btns(root):
    myFont = font.Font(size=35)
    button=[]
    for i in range(3):
            for j in range(3):
                val = (3*i)+j+1
                col='blue'
                button.append(tk.Button(root,text=str(val),bg=col))
                button[val-1]['font'] = myFont
                button[val-1].grid(row=i+1,column=j,sticky=tk.W)
    return button
                
#root.mainloop()
def lbls(root,vals):
    myFont = font.Font(size=30)
    l1 = tk.Label(root, text = vals)
    l1.grid(row=0,column=1,sticky=tk.W)
    l1['font'] = myFont
def chngcol(button):
    button.configure(background = 'green')
```

```python
cap = cv2.VideoCapture(0)
#cap = cv2.VideoCapture('http://192.168.0.100:8080/video')
root = tk.Tk()
vals=''
frs=0
buttons=btns(root)
while True:
    frs+=1
    _, frame = cap.read()
    cv2.imshow("frame", frame)
    key = cv2.waitKey(1)
    # reading the image 
    panel = tk.Label(root, image = frame) 
  
    # setting the application 
    panel.pack(side = "bottom", fill = "both", 
           expand = "yes") 
    if frs==34:
        chngcol(buttons[0])
    elif frs==145:
        chngcol(buttons[5])
    try:
        lbls(root,vals)
        root.update()
    except:
        break
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
